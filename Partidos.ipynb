```python
matches["time"] = pd.to_datetime(matches["time"], format="%H:%M")
matches["hour"] = matches["time"].dt.hour
```

```python
matches["venue_code"] = matches["venue"].astype("category").cat.codes
matches
```

```python
matches["opp_code"] = matches["opponent"].astype("category").cat.codes
matches
```

```python
matches["day_code"] = pd.to_datetime(matches.index).dayofweek
matches
```

```python
matches["target"] = (matches["result"] == "W").astype(int)
matches
```

```python
from sklearn.ensemble import RandomForestClassifier 
```

```python
rf = RandomForestClassifier(n_estimators=100, min_samples_split=2, random_state=1)
```

```python
train = matches[matches.index < "2026-01-01"]
```

```python
test = matches[matches.index >= "2026-01-01"]
```

```python
predictors = ["venue_code", "opp_code", "hour", "day_code"]
```

```python
rf.fit(train[predictors], train["target"])
```

```python
RandomForestClassifier(min_samples_split=10, n_estimators=50, random_state=1)
```

```python
preds = rf.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(test["target"], preds)
```

```python
def rolling_average(group, cols, new_cols):
    group = group.sort_index().copy()
    group[new_cols] = group[cols].rolling(3, closed="left").mean()
    return group.dropna(subset=new_cols)

cols = ["gf", "ga", "sh", "sot", "fouls", "corners", "yellow", "red"]
new_cols = [f"{col}_rolling" for col in cols]

rolling_average(group, cols, new_cols)

```
